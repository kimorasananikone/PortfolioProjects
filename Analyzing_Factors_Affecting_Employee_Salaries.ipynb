```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Salary Data.csv')

#Data Exploration
print("Dataset Overview:")
print(data.head())
print(data.info())
```

```python
# Data Cleaning
# Handle missing values
data = data.dropna(subset=['salary','experience','education_level'])

# Standarizing catagories
# Converting all salary values to common currency and adjusting education levels to standard categories.
data['education_level'] = data['education_level'].replace({
    'Bachelors':'Bachelor’s',
    'Masters':'Master’s',
    'phd':'PhD'
})
```

```python
# Exploratory Data Analysis
# Salary Distribution

plt.figure(figsize=(10,6))
sns.histplot(data['salary'], kde = True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average Salary by Education Level
avg_salary_by_education = data.groupby('education_level')['salary'].mean()
plt.figure(figsize=(10,6))
avg_salary_by_education.plot(kind='bar',color='skyblue')
plt.title('Average Salary by Education Level')
plt.xlabel("Education Level")
plt.ylabel("Average Salary")
plt.show()
```

```python
# Salary vs Years of Experience
plt.figure(figsize=(10,6))
sns.lineplot(x='experience',y='salary',data=data, marker='o',ci=None)
plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Average Salary by Job Title
avg_salary_by_job = data.groupby('Job Title')['salary'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
avg_salary_by_job.plot(kind='bar',color='purple')
plt.title("Top 10 Job Titles by Average Salary")
plt.xlabel("Job Title")
plt.ylabel("Average Salary")
plt.show()
```
